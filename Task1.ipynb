```python
import pandas as pd
df=pd.read_csv("healthcare_dataset (1).csv")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# ✅  BAR CHART for Categorical Variable (Gender)

Gender_data=df["Gender"].value_counts()

plt.figure(figsize=(6,4))
Ax=Gender_data.plot(kind="bar",color=["blue","pink"])
Ax.bar_label(Ax.containers[0])
plt.title("Distribution of Gender")
plt.show()
```

```python
# ✅  HISTOGRAM for Continuous Variable (Age)
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], kde=True, bins=5, color='skyblue')  # kde=True adds smooth curve
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Bar chart → shows how many males & females.

Histogram → shows how ages are spread (more people in 40-60 range, etc.).
